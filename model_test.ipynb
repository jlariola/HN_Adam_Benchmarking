```python
import torch
import torch.nn as nn

# 1. Check if the GPU is available to PyTorch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"PyTorch is using device: {device}")

# 2. Create a very simple, blank neural network
class SimpleModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(10, 2)

model = SimpleModel()

# 3. Send the model to the GPU
model.to(device)

# 4. Verify the model is actually on the GPU
print(f"Model is located on: {next(model.parameters()).device}")
```
